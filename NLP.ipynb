```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
X = df.text
```

```python
vectorizer = TfidfVectorizer()
```

```python
print(df.text.loc[4])
```

```python
XX = vectorizer.fit_transform(X)
tfidf_df = pd.DataFrame(XX.toarray(), columns = vectorizer.get_feature_names())
W = XX.toarray()
```

```python
tfidf_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(tfidf_df)
df_similarity = pd.DataFrame(similarity, index=tfidf_df.index, columns=tfidf_df.index)
df_similarity.head()
```

```python
def calc_similarity(txt):
    new = tfidf_df[1000:]
    for word in txt.split():
        if word in vectorizer.get_feature_names():
            new.loc[0, word] = 1
    new.fillna(0, inplace = True)
    new_similarity = cosine_similarity(tfidf_df, new)
    return np.argmax(new_similarity)
```

```python
calc_similarity('شخصیت ایرانی اسلام')
```
